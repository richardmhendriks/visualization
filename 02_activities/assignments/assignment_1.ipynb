Making a basic figure with matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# insert a random seed to create reproducibility
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)
```

```python
# define the plot types and sizes for scatter visualization
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
# define the plot types and sizes for bar chart visualization
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
#modify to "barh" to make it a horizontal bar chart
fig, ax = plt.subplots(figsize=(5, 3))
ax.barh(x,y)
```

```python
# define the plot types and sizes for line graph visualization
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
# define the plot types and sizes for histogram visualization
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
# insert axis titles to improve interpretation
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
# insert axis titles to improve interpretation
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
# alter fonts to improve presentation
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
# use tight layout to 
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'center')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

Text and Annotations

```python
# adding a title to a scatter plot
fig, ax  = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = 'Person 1')
ax.scatter(x,y2, label = 'Person 2')
ax.legend(loc='lower right')
ax.text(10,95, "This value is important!",
        ha = 'center',
        color = 'red',
        size = 20)
fig.show()
```

```python
# alter text start location, which is based on the x, y axis
fig, ax  = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = 'Person 1')
ax.scatter(x,y2, label = 'Person 2')
ax.legend(loc='lower right')
ax.text(20,85, "This value is important!")
fig.show()
```

```python
# positioning text annotations
fig, ax  = plt.subplots()
# the square brackets defines the range, which must be in square brackets
ax.axis([0,10,0,10])
ax.text(1,5, ". Data:(1,5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure)
fig.show()
```

```python
# sub-plots without a grid arrangement
# we can use subplot_mosaic to arrange subplots
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], ['ax2', 'ax3']], figsize=(7, 4))
for label, ax in someaxes.items():
        ax.text(0.5, 0.5, f'This is {label!r}', 
        fontsize=14, 
        ha = 'center', 
        transform=ax.transAxes)
```

```python
# now we can set the type of each subplot
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], ['ax2', 'ax3']], figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

Model figure layout

```python
# Tight layout adjusts subplots so tick labels, axis labels, and titles don't overlap or leave the figure area
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], ['ax2', 'ax3']], figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
fig.tight_layout()
plt.show()
```

```python
# Constrained layout works similarly except it also fits things like legends or colorbars
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], ['ax2', 'ax3']], figsize=(7, 4), layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
plt.show()
```

Multiple Visualizations on One Axis

```python
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
```

```python
# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))
```

```python
# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1, color = "indigo")
ax.plot(x, y2, color = "red")
```

```python
#then, calculate the standard deviation
y2_sd = np.std(y2)
```

```python
# then plot the lines as before
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
```

```python
# use error bar to add in an error line
# yerr specifies that we’re plotting vertical error bars
# fmt makes sure we’re not plotting the actual data points, only the error

ax.errorbar(x, #our x values
            y2, #our y values
            yerr = y2_sd, fmt = "none")
```

```python
#Customize errobar appearance
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
ax.errorbar(x, y2, yerr = y2_sd, fmt = "none", ecolor= "indigo", elinewidth= 4, capsize = 6, capthick= 4 )
```

```python
#If we don’t want to see error bars for every single point, we can specify intervals using errorevery
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
ax.errorbar(x, y2, yerr = y2_sd, fmt = "none", ecolor = "indigo", elinewidth = 4, capsize = 6, capthick = 4, errorevery= 2 )
```

Making Venn Diagrams

```python
#We can use the matplotlib_venn extension to make venn diagrams
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
# Define our sets
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```

```python
# assign our sets to each circle and modify the appearance

diagram = venn2_unweighted([A, B],
                           set_labels = ('Fruits', 'Pies'),
                           set_colors=("blue", "red"),
                           alpha=0.5)
plt.show()
```

```python
diagram = venn2_unweighted([A, B],
                           set_labels = ('Fruits', 'Pies'),
                           set_colors=("blue", "red"),
                           alpha=0.5)
# \ places the text on a new line
diagram.get_label_by_id("10") .set_text("\n".join(A - B))
diagram.get_label_by_id("11") .set_text("\n".join(A & B))
diagram.get_label_by_id("01") .set_text("\n".join(B - A))
plt.show()
```
